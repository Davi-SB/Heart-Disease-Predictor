```python
import pandas as pd
from IPython.display import display
```

```python
SHOW = 10

# - Definindo o cabeçalho que será usado para todos os datasets
headerList = ['age','sex','chestPain','restingBloodPressure','serumCholestoral','fastingBloodSugar>120','restingEletroc',
              'maximumHeartRate','exerciseInducedAngina','oldpeak','op_slope','numberMajorVessels','bloodFlow_thal','diagnosis']
    
def show_df(df):
    print('\n---------------------\n')
    display(df.head(SHOW))
```

```python
# - Lendo os arquivos CSV sem cabeçalho e atribuindo o cabeçalho definido
# O parâmetro header=None informa que o arquivo não possui cabeçalho e o parâmetro names
# atribui o cabeçalho customizado que definimos
df1 = pd.read_csv('data/processed.cleveland.csv',   header=None, names=headerList)
df2 = pd.read_csv('data/processed.hungarian.csv',   header=None, names=headerList)
df3 = pd.read_csv('data/processed.switzerland.csv', header=None, names=headerList)
df4 = pd.read_csv('data/processed.va.csv',          header=None, names=headerList)
```

```python
show_df(df1)
show_df(df2)   
show_df(df3)
show_df(df4)
```

```python

# Removendo as colunas indesejadas de cada dataframe
colunas_remover = ['fastingBloodSugar>120','oldpeak', 'op_slope','numberMajorVessels','bloodFlow_thal']
df1 = df1.drop(columns=colunas_remover)
df2 = df2.drop(columns=colunas_remover)
df3 = df3.drop(columns=colunas_remover)
df4 = df4.drop(columns=colunas_remover)
```

```python
show_df(df1)
show_df(df2)   
show_df(df3)
show_df(df4)
```

```python
# Cast de todas as colunas de df1 para int
df1 = df1.astype(int)
```

```python
# Remove linhas com caractere "?" de todos os dataframes
df1 = df1[(df1 != '?').all(1)]
df2 = df2[(df2 != '?').all(1)]
df3 = df3[(df3 != '?').all(1)]
df4 = df4[(df4 != '?').all(1)]
```

```python
show_df(df1)
show_df(df2)   
show_df(df3)
show_df(df4)
```

```python
# Substitui os valores 1 e 2 por 0, e os valores 3 e 4 por 1 da coluna diagnosis nos dataframes necessários
df1['diagnosis'] = df1['diagnosis'].replace([1, 2], 0)
df1['diagnosis'] = df1['diagnosis'].replace([3, 4], 1)

df3['diagnosis'] = df3['diagnosis'].replace([1, 2], 0)
df3['diagnosis'] = df3['diagnosis'].replace([3, 4], 1)

df4['diagnosis'] = df4['diagnosis'].replace([1, 2], 0)
df4['diagnosis'] = df4['diagnosis'].replace([3, 4], 1)
```

```python
show_df(df1)
show_df(df2)   
show_df(df3)
show_df(df4)
```

```python
# exporta os dataframes para arquivos CSV
df1.to_csv('data/view.cleveland.csv', index=False)
df2.to_csv('data/view.hungarian.csv', index=False)
df3.to_csv('data/view.switzerland.csv', index=False)
df4.to_csv('data/view.va.csv', index=False)
```

```python
# Junta os 4 dataframes em um único e exporta para um arquivo CSV com cabeçalho
df = pd.concat([df1, df2, df3, df4])
df = df.astype(int)
df.to_csv('data/all.csv', index=False)

show_df(df)

# Exbindo informações do dataframe final
print("\nInformações do dataframe final:")
df.info()
```
